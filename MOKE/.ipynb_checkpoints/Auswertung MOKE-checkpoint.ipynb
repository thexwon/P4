```python
import numpy as np
import matplotlib.pyplot as plt
import glob

def read_data(path):
    """returns B in mT and Kerr rotation in mdeg"""
    x = np.loadtxt(path,delimiter = "\t")
    return x[:,2], x[:,6]

def plot_data(B, Kerr,label):
    plt.scatter(B,Kerr,marker=".",color="black",label=label)
    plt.grid()
    plt.xlabel("B in mT")
    plt.ylabel("Kerr rotation in m°")
    plt.axhline(0, color='orange', linestyle='--')
    plt.axvline(0,color="gray",linestyle="--")

def center_data(kerr):
    return kerr - np.mean(kerr)
    
def plot_centered(B,Kerr):
    Kerr_centered = Kerr - np.mean(Kerr)
    
    plt.scatter(B, Kerr_centered,marker="x",color="black")
    plt.grid()
    plt.xlabel("B in mT")
    plt.ylabel("Centered Kerr rotation in m°")
    plt.axhline(0, color='gray', linestyle='--')  # add a zero line
```

```python
b, kerr = read_data("Daten/sample6_lmoke_a3_i3")
plot_data(b,kerr,"")
```

```python
# task 1 & 2:
samples = ["sample6", "sample7", "sample8"]
setups = ["lmoke", "pmoke"]

fig, axes = plt.subplots(3, 2, figsize=(10, 9))

for i in range(3):
    # sharey only within the same row
    axes[i, 1].sharey(axes[i, 0])
for i, sample in enumerate(samples):
    for j, setup in enumerate(setups):
        ax = axes[i, j]
        path = f"Daten/{sample}_{setup}_a3_i3"
        B, Kerr = read_data(path)
        centered_kerr = center_data(Kerr)
        # Plot
        
        ax.scatter(B, centered_kerr, s=10)
        ax.axhline(0, color='black', linestyle='--', linewidth=0.8)
        ax.axvline(0, color='orange', linestyle='--', linewidth=0.8)
        ax.grid()
        # Titles and labels
        ax.set_title(f"{sample.upper()} - {setup.upper()}")
        if j == 0:
            ax.set_ylabel("Kerr (m°)")
        if i == 2:
            ax.set_xlabel("B (mT)")

plt.tight_layout()
plt.savefig("task12.png",dpi=300)
plt.show()
```

```python

# === Task 3: single hysteresis for each Fe sample (#1, #2, #4) ===
samples = ["sample1", "sample2", "sample4"]

fig, axes = plt.subplots(1, 3, figsize=(12, 4), sharey=True)
for i, sample in enumerate(samples):
    path = f"Daten/task3_{sample}_lmoke_a3_i3"
    B, Kerr = read_data(path)
    Kerr_c = center_data(Kerr)
    axes[i].scatter(B, Kerr_c, s=10, color="black")
    axes[i].axhline(0, color="gray", linestyle="--", linewidth=0.8)
    axes[i].axvline(0, color="orange", linestyle="--", linewidth=0.8)
    axes[i].set_title(f"{sample.upper()} – l-MOKE")
    axes[i].set_xlabel("B (mT)")
    if i == 0:
        axes[i].set_ylabel("Kerr (m°)")
    axes[i].grid(True)

plt.tight_layout()
plt.savefig("task3_Fe_lMOKE.png", dpi=300)
plt.show()
```

```python
# === Task 4: angular dependence for sample #4 ===
# automatically load all files like task4_sample4_*.txt
paths = sorted(glob.glob("Daten/task4_sample4_*"))
angles = []
fig, ax = plt.subplots(figsize=(6, 5))

for path in paths:
    # try to extract angle number from filename (e.g. 0, 45, 90, 135, 180)
    try:
        angle = int(path.split("_")[-1])
    except ValueError:
        angle = None
    B, Kerr = read_data(path)
    Kerr_c = center_data(Kerr)
    #plot_data(B, Kerr_c, label=f"{angle}°" if angle is not None else path)
    if angle is not None:
        angles.append(angle)
    plt.scatter(B,Kerr_c,marker=".",label=angle)
    plt.grid()
    plt.axhline(0, color='orange', linestyle='--')
    plt.axvline(0,color="gray",linestyle="--")
    plt.xlabel("B (mT)")
    plt.ylabel("Centered Kerr (m°)")
    plt.title("Sample #4 – l-MOKE at different angles α")
    plt.legend(title="α")
    plt.tight_layout()
    plt.show()
#plt.savefig("task4_Fe_lMOKE_angles.png", dpi=300)
```

```python
H_c = [17.5,4.5,18.5,10,14,19,13,8,7.5,23.5,5.5,12,17.5]
alpha = np.array([0,30,60,90,120,150,180,210,240,270,300,330,360])


# Convert degrees to radians
theta = np.deg2rad(alpha)

# Create polar plot
plt.figure(figsize=(6,6))
ax = plt.subplot(111, polar=True)
ax.plot(theta, H_c, marker='o', linestyle='-')
ax.set_theta_zero_location("N")  # 0 degrees at the top
ax.set_theta_direction(-1)       # Clockwise direction
plt.title("Angular dependence of the coercivity")
plt.savefig("task4.png",dpi=300)
plt.show()
```
