```python
import numpy as np
import scipy.optimize as sp
import matplotlib.pyplot as plt
```

```python
from scipy.optimize import curve_fit


def gauss_bg(x, A1, x01, sigma1):
    g1 = A1 * np.exp(-(x - x01)**2 / (2 * sigma1**2))
    return g1 
```

```python
ge_calibration  = []
nai_calibration = []

# sigma lists for energy-resolution analysis (channel-space, from Gaussian fits)
ge_sigma_ch  = []   # (E_keV, sigma_channel) for Ge calibration peaks
nai_sigma_ch = []   # (E_keV, sigma_channel) for NaI calibration peaks
```

```python
data  = np.load("data/Co57_Ge/Histograms_PulseHeights/height.npy")
p0 = [3400,268 , 20]
theoretical = 122
lim1 =263
lim2 =278
x = np.arange(len(data))
plt.scatter(x,data, color = "black", label = "data", marker='x')
plt.xlim(lim1-100, lim2+100)
plt.axvline(lim1, color = "gray", lw=0.5,ls="-.")
plt.axvline(lim2, color = "gray", lw=0.5,ls="-.")

popt, cov = curve_fit(gauss_bg,np.arange(len(data))[lim1:lim2], data[lim1:lim2], p0 = p0)

plt.plot(x, gauss_bg(x,*popt), label = "fit", color = "red", alpha = 0.8)
plt.legend()

plt.show()
print(popt)


ge_calibration.append([popt[1], theoretical])

ge_sigma_ch.append((theoretical, popt[2]))   # store sigma_channel for resolution analysis
```

```python
def line(x, a, b):
    return a*x + b
```

```python
x = np.array([row[0] for row in ge_calibration])  # erste Spalte (float)
y = np.array([row[1] for row in ge_calibration])  # zweite Spalte (int)

popt, pcov = curve_fit(line, x, y)

plt.scatter(x, y, marker = 'x', color = 'black', label='data')
plt.plot(x, line(x, *popt), color = 'red', label = 'linear fit')
plt.xlabel('peak position')
plt.ylabel('theoretical value')
plt.legend()
plt.grid(True)
plt.savefig('linefit_ge.jpg', dpi = 800)
plt.show()
print('y-axis offset: ', popt[1])

aGe = popt[0]
bGe = popt[1]

def EGe(x):
    return aGe*x + bGe
```

```python
x = np.array([row[0] for row in nai_calibration])  # erste Spalte (float)
y = np.array([row[1] for row in nai_calibration])  # zweite Spalte (int)

popt, pcov = curve_fit(line, x, y)

plt.scatter(x, y, marker = 'x', color = 'black', label='data')
plt.plot(x, line(x, *popt), color = 'red', label = 'linear fit')
plt.xlabel('peak position')
plt.ylabel('theoretical value')
plt.legend()
plt.grid(True)
plt.savefig('linefit_nai.jpg', dpi = 800)
plt.show()
print('y-axis offset: ', popt[1])


aNaI = popt[0]
bNaI = popt[1]

def ENaI(x):
    return aNaI*x + bNaI
```

```python
from scipy.optimize import minimize_scalar
import matplotlib.gridspec as gridspec

N_measured = counts
print(counts)
#N_measured = np.array([21, 63, 64, 3])   #default values from some pasta protocoll
NG = np.sum(N_measured)                   

N_corr = np.array([0, NG/2, NG/2, 0])

N_uncorr = np.array([NG/4, NG/4, NG/4, NG/4])


def chi2(r):
    Nm = r * N_corr + (1 - r) * N_uncorr
    # Avoid division by zero (Nm=0 only when correlated term=0 AND alpha=1)
    safe = np.where(Nm > 0, Nm, 1e-10)
    return np.sum((N_measured - Nm)**2 / safe)


result = minimize_scalar(chi2, bounds=(0, 1), method='bounded')
r_min = result.x
chi2_min  = result.fun


print(f"  chi² minimisation result:")
print(f"    r_min  = {r_min:.6f}  ({r_min*100:.1f}% correlated)")
print(f"    chi²_min = {chi2_min:.6f}")

# Compare pure hypotheses
print(f"\n  χ²(r=0)  [fully uncorrelated] = {chi2(0):.4f}")
print(f"  χ²(r=1)  [fully correlated]   = {chi2(1):.4f}")
print(f"  χ²(r_min)[best fit]           = {chi2_min:.4f}")



fig = plt.figure(figsize=(13, 9))
gs = gridspec.GridSpec(2, 2, figure=fig, hspace=0.38, wspace=0.32)

# ── Plot 1: chi^2 as a function of alpha ──────────────────
ax1 = fig.add_subplot(gs[0, :])
alphas = np.linspace(0, 0.9, 500)
chi2_vals = [chi2(a) for a in alphas]
ax1.plot(alphas, chi2_vals, color='steelblue', lw=2, label=r'$\chi^2(r)$')
ax1.axvline(r_min, color='crimson', ls='--', lw=1.5,
            label=rf'$r_{{min}}={r_min:.3f}$  ($\chi^2={chi2_min:.2f}$)')
ax1.axhline(chi2_min, color='crimson', ls=':', lw=1, alpha=0.6)
ax1.set_xlabel(r'Mixing parameter $\alpha$  (0 = uncorrelated, 1 = fully correlated)', fontsize=11)
ax1.set_ylabel(r'$\chi^2(\alpha)$', fontsize=11)
ax1.set_title(r'$\chi^2$ test: Co-60 coincidence – correlated vs. uncorrelated emission', fontsize=12)
ax1.legend(fontsize=10)
ax1.grid(True, alpha=0.3)

# ── Plot 2: Measured vs. expected counts ──────────────────
ax2 = fig.add_subplot(gs[1, 0])
window_labels = ['N₁\n(NaI-1,Ge-1)', 'N₂\n(NaI-1,Ge-2)',
                 'N₃\n(NaI-2,Ge-1)', 'N₄\n(NaI-2,Ge-2)']
x = np.arange(4)
w = 0.25
N_best = r_min * N_corr + (1 - r_min) * N_uncorr
ax2.bar(x - w, N_measured, w, color='steelblue',  label='Measured')
ax2.bar(x,     N_corr,     w, color='tomato',     label='Fully correlated', alpha=0.8)
ax2.bar(x + w, N_uncorr,   w, color='seagreen',   label='Fully uncorrelated', alpha=0.8)
ax2.errorbar(x - w, N_measured, yerr=np.sqrt(N_measured),
             fmt='none', color='black', capsize=4, lw=1.5)
ax2.set_xticks(x)
ax2.set_xticklabels(window_labels, fontsize=9)
ax2.set_ylabel('Counts', fontsize=11)
ax2.set_title('Counts per coincidence window', fontsize=11)
ax2.legend(fontsize=8)
ax2.grid(True, axis='y', alpha=0.3)

# ── Plot 3: Measured vs. best-fit model ───────────────────
ax3 = fig.add_subplot(gs[1, 1])
ax3.bar(x - w/2, N_measured, w, color='steelblue', label='Measured')
ax3.bar(x + w/2, N_best,     w, color='darkorange', alpha=0.85,
        label=rf'Best fit ($r={r_min:.2f}$)')
ax3.errorbar(x - w/2, N_measured, yerr=np.sqrt(N_measured),
             fmt='none', color='black', capsize=4, lw=1.5)
ax3.set_xticks(x)
ax3.set_xticklabels(window_labels, fontsize=9)
ax3.set_ylabel('Counts', fontsize=11)
ax3.set_title('Measured vs. best-fit model', fontsize=11)
ax3.legend(fontsize=9)
ax3.grid(True, axis='y', alpha=0.3)
plt.savefig("coincidence.png")
```

## Energy Resolution

```python
# ─────────────────────────────────────────────────────────────
#  Energy Resolution
# ─────────────────────────────────────────────────────────────
# ge_sigma_ch  / nai_sigma_ch contain (E_keV, sigma_channel)
# collected directly from popt[2] of every earlier Gaussian fit.
# Convert to keV using the calibration slopes, then compute
#   FWHM = 2*sqrt(2*ln2) * sigma_keV
#   Delta_E/E = FWHM / E

FWHM_factor = 2 * np.sqrt(2 * np.log(2))   # ~2.355

def build_rows(sigma_ch_list, slope, co60_pairs):
    """Return sorted list of (E_keV, sigma_keV, FWHM_keV, dE_over_E)."""
    rows = []
    for E, sigma_ch in sigma_ch_list:
        sigma_keV = slope * abs(sigma_ch)
        FWHM      = FWHM_factor * sigma_keV
        rows.append((E, sigma_keV, FWHM, FWHM / E))
    for E, sigma_keV in co60_pairs:
        sigma_keV = abs(sigma_keV)
        FWHM      = FWHM_factor * sigma_keV
        rows.append((E, sigma_keV, FWHM, FWHM / E))
    return sorted(rows, key=lambda r: r[0])

# Co60 sigmas are already in keV (computed in cells 18/19 and 22/23)
ge_rows  = build_rows(ge_sigma_ch,  aGe,
                      [(Co60_Ge_peak1,  Co60_Ge_sigma1),
                       (Co60_Ge_peak2,  Co60_Ge_sigma2)])
nai_rows = build_rows(nai_sigma_ch, aNaI,
                      [(Co60_NaI_peak1, Co60_NaI_sigma1),
                       (Co60_NaI_peak2, Co60_NaI_sigma2)])

header = f"{'E (keV)':>10}  {'sigma (keV)':>12}  {'FWHM (keV)':>12}  {'dE/E':>8}"
fmt    = "{:10.1f}  {:12.4f}  {:12.4f}  {:8.5f}"

print("=== Ge Detector ===")
print(header)
for r in ge_rows:
    print(fmt.format(*r))

print()
print("=== NaI Detector ===")
print(header)
for r in nai_rows:
    print(fmt.format(*r))
```
