```python
def n(lam):
    return 1.94 * (1 + 4696/(2*sp.constants.pi*sp.constants.c)/(1/(390)**2 - 1/(lam)**2))


def line(x,a,b):
    return 4*a*x + b

plt.scatter(n(peaks_red)/peaks_red, np.arange(0, 5, 1), color = 'red',
           label = 'minimum transmission peaks')
popt_red, pcov_red = sp.optimize.curve_fit(line, n(peaks_red)/peaks_red, np.arange(0, 5, 1))
plt.plot(n(peaks_red)/peaks_red, line(n(peaks_red)/peaks_red, *popt_red),
         color = 'maroon', linestyle = '--', 
         label = fr'fit: $d = {np.round(-popt_red[0], 0)} \pm {np.round(pcov_red[0][0]**0.5, 0)}$ nm')
plt.scatter(n(peaks_blue)/peaks_blue, np.arange(0, 5, 1), color = 'blue', 
            label = 'maximum transmission peaks')
popt_blue, pcov_blue = sp.optimize.curve_fit(line, n(peaks_blue)/peaks_blue, np.arange(0, 5, 1))
plt.plot(n(peaks_blue)/peaks_blue, line(n(peaks_blue)/peaks_blue, *popt_blue), 
         color = 'navy', linestyle = '--', 
        label = fr'fit: $d = {np.round(-popt_blue[0], 0)} \pm {np.round(pcov_blue[0][0]**0.5, 0)}$ nm')

plt.xlabel(fr'$\frac{{4n(\lambda)}}{{n}}$')
plt.ylabel('peak order')
plt.grid()
plt.legend()
plt.savefig('ss_ex2_linear_fits.jpg', dpi = 800)

d_par = ufloat(-popt_red[0], pcov_red[0][0]**0.5)
d_ort = ufloat(-popt_blue[0], pcov_blue[0][0]**0.5)
d = (d_par+d_ort)/2
print(fr'd = {d:u.4}')
```

```python
start = 850
stop = 400
plt.plot(readd('Aufg4_Dick_Kalt.Master.Absorbance', start, stop)[:,0], 
        readd('Aufg4_Dick_Kalt.Master.Absorbance', start, stop)[:,1],
        label = 'absorbance at 77K',
         color = 'blue'
        )

lambda3p = ufloat(578, 3)
lambda2p = ufloat(583, 3)
lambda1s_1 = ufloat(613, 3)
lambda1s_2 = ufloat(622, 3)
lambda1s = (lambda1s_1+lambda1s_2)/2

plt.axvline(x = lambda3p.nominal_value, color = 'green', linestyle = '--', 
            label = fr'3p ($\lambda = {lambda3p.nominal_value:.0f}$ nm)')
plt.axvline(x = lambda2p.nominal_value, color = 'red', linestyle = '--', 
            label = fr'2p ($\lambda = {lambda2p.nominal_value:.0f}$ nm)')
plt.axvline(x = lambda1s_1.nominal_value, color = 'black', linestyle = '--', 
            label = fr'1s ($\lambda = \frac{{{lambda1s_2.nominal_value:.0f} - {lambda1s_1.nominal_value:.0f}}}{{2}}$ nm = {lambda1s.nominal_value:.1f} nm)')
plt.axvline(x = lambda1s_2.nominal_value, color = 'black', linestyle = '--')

plt.xlabel(fr'$\lambda$ [nm]')
plt.ylabel(fr'$I$')
plt.grid()
plt.legend()
plt.savefig('ss_ex4_thick_cold.jpg', dpi = 800)

print(fr'lambda3p = {lambda3p} nm')
print(fr'lambda2p = {lambda2p} nm')
print(fr'lambda1s_1 = {lambda1s_1} nm')
print(fr'lambda1s_2 = {lambda1s_2} nm')
print(fr'lambda1s = {lambda1s} nm')

E3p = c.h*c.c/lambda3p*1e9
E2p = c.h*c.c/lambda2p*1e9
E1s = c.h*c.c/lambda1s*1e9

print(fr'E3p = {E3p/c.e:u.4} eV')
print(fr'E2p = {E2p/c.e:u.4} eV')
print(fr'E1s = {E1s/c.e:u.4} eV')

E_g = 4/3*E3p - 1/3*E2p
print(fr'E_g(2p-1s) = {E_g/c.e:u.4} eV')

E_g = 9/5*E3p - 4/5*E2p
print(fr'E_g(3p-2p) = {E_g/c.e:u.4} eV')

E_b = 4/3*(E2p - E1s)
print(fr'E_b(2p-1s) = {E_b/c.e:u.4} eV')

E_b = 1/(1/4-1/9)*(E3p - E2p)
print(fr'E_b(3p-2p) = {E_b/c.e:u.4} eV')
```
